### Lambda Functions -: 

Lambda functions are small anonymous functions defined using the lambda keyword. They can have number of arguments but only one expression. They are commonly used for short operation or as arguments to higher- order functions.

Basically a function without a name.

```python
## Syntax for writing a lambda function 

lambda arguments: expression 
```

```python
##For Example

def addition(a,b):
    return a+b

print(addition(2,3))

## Defining lambda function

add=lambda a,b: a+b

print(add(6,3))
```

```python
## Find even number using lambda 

even= lambda a: a%2==0

even(9)
```

```python
## Lambda function for multiple arguments

sub= lambda a,b,c: a-b-c

sub(20,10,5)
```

```python
def even(num):
    if num%2==0:
        return True

even(10)
```

```python
even1=lambda n:n%2==0

even1(8)
```
